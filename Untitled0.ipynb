```python
# Import libraries
import os
import shutil
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from torchvision.models import resnet18
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
```

```python
# Set random seeds for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
torch.cuda.manual_seed_all(42)
```

```python
# Setup paths (modify these to your dataset location in Drive)
base_dir = '/content/drive/MyDrive/result'
os.makedirs(base_dir, exist_ok=True)
```

```python
# Define original directories (update these paths)
original_dirs = {
    'Calculus': os.path.join(base_dir, 'Calculus'),
    'Gingivitis': os.path.join(base_dir, 'Gingivitis')
}

# Create directory structure
splits = ['train', 'val', 'test']
# Explicitly define the classes being used
classes = ['Calculus', 'Gingivitis']

for split in splits:
    for class_name in classes:
        os.makedirs(os.path.join(base_dir, split, class_name), exist_ok=True)
```

```python
# Create directory structure
splits = ['train', 'val', 'test']
classes = list(original_dirs.keys())

for split in splits:
    for class_name in classes:
        os.makedirs(os.path.join(base_dir, split, class_name), exist_ok=True)
```

```python
# Function to organize dataset
def organize_dataset():
    class_split_counts = {class_name: {'train': 0, 'val': 0, 'test': 0} for class_name in classes}

    for class_name, original_dir in original_dirs.items():
        image_paths = []
        for root, _, files in os.walk(original_dir):
            image_paths.extend([os.path.join(root, f) for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

        if image_paths:
            # Split the data
            train_paths, test_paths = train_test_split(image_paths, test_size=0.1, random_state=42)
            train_paths, val_paths = train_test_split(train_paths, test_size=0.2, random_state=42)

            # Copy images
            for split, paths in zip(['train', 'val', 'test'], [train_paths, val_paths, test_paths]):
                class_split_counts[class_name][split] = len(paths)
                for img_path in paths:
                    shutil.copy(img_path, os.path.join(base_dir, split, class_name, os.path.basename(img_path)))

    return class_split_counts
```

```python
# Organize the dataset
class_split_counts = organize_dataset()
print("Dataset organized successfully!")
```

```python
# Visualize class distribution
def plot_class_distribution():
    labels = classes
    train_counts = [class_split_counts[c]['train'] for c in classes]
    val_counts = [class_split_counts[c]['val'] for c in classes]
    test_counts = [class_split_counts[c]['test'] for c in classes]

    plt.figure(figsize=(14, 7)) # Increased figure size
    plt.bar(labels, train_counts, label='Train')
    plt.bar(labels, val_counts, bottom=train_counts, label='Val')
    plt.bar(labels, test_counts, bottom=[i+j for i,j in zip(train_counts, val_counts)], label='Test')
    plt.xticks(rotation=45, ha='right') # Rotated and aligned labels
    plt.ylabel('Number of Images')
    plt.title('Dataset Distribution')
    plt.legend()
    plt.tight_layout() # Adjust layout to prevent labels overlapping
    plt.show()

plot_class_distribution()
```

```python
# Define transforms
image_size = (128, 128)
batch_size = 32

train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(image_size, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])

val_transforms = transforms.Compose([
    transforms.Resize(image_size),
    transforms.CenterCrop(image_size),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])
```

```python
# Create datasets
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')

train_dataset = ImageFolder(train_dir, transform=train_transforms)
val_dataset = ImageFolder(val_dir, transform=val_transforms)
test_dataset = ImageFolder(test_dir, transform=val_transforms)
```

```python
# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
```

```python
# Define model
class OralDiseaseModel(nn.Module):
    def __init__(self, num_classes):
        super(OralDiseaseModel, self).__init__()
        self.model = resnet18(weights='IMAGENET1K_V1')
        self.model.fc = nn.Linear(self.model.fc.in_features, num_classes)

    def forward(self, x):
        return self.model(x)
```

```python
# Initialize model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# Get the number of classes from the training dataset
num_classes = len(train_dataset.classes)
model = OralDiseaseModel(num_classes=num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=2, verbose=True)
```

```python
# Training function
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=25):
    best_val_loss = float('inf')
    patience = 3
    epochs_no_improve = 0

    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        epoch_loss = running_loss / len(train_dataset)
        epoch_acc = correct / total
        train_losses.append(epoch_loss)
        train_accs.append(epoch_acc)

        # Validation
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        val_loss = val_loss / len(val_dataset)
        val_acc = val_correct / val_total
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        print(f'Epoch {epoch+1}/{num_epochs}')
        print(f'Train Loss: {epoch_loss:.4f} | Val Loss: {val_loss:.4f}')
        print(f'Train Acc: {epoch_acc:.4f} | Val Acc: {val_acc:.4f}')

        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            epochs_no_improve = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print('Early stopping!')
                break

        scheduler.step(val_loss)

    return train_losses, val_losses, train_accs, val_accs
```

```python
# Train the model
train_losses, val_losses, train_accs, val_accs = train_model(
    model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=20)
```

```python
# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.legend()
plt.title('Loss Curves')

plt.subplot(1, 2, 2)
plt.plot(train_accs, label='Train Acc')
plt.plot(val_accs, label='Val Acc')
plt.legend()
plt.title('Accuracy Curves')
plt.show()
```

```python
# Evaluation
model.eval()
y_true = []
y_pred = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())
```

```python
# Classification report
print(classification_report(y_true, y_pred, target_names=classes))
```

```python
# Confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
# Grad-CAM visualization
def visualize_grad_cam(model, dataloader, classes, num_images=6):
    target_layer = model.model.layer4[-1]
    cam = GradCAM(model=model, target_layers=[target_layer])

    plt.figure(figsize=(15, 10))
    images_so_far = 0

    for inputs, labels in dataloader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        for i in range(inputs.size(0)):
            if images_so_far >= num_images:
                return

            input_tensor = inputs[i].unsqueeze(0)
            grayscale_cam = cam(input_tensor=input_tensor)
            grayscale_cam = grayscale_cam[0, :]

            rgb_img = inputs[i].cpu().permute(1, 2, 0).numpy()
            rgb_img = (rgb_img - rgb_img.min()) / (rgb_img.max() - rgb_img.min())
            visualization = show_cam_on_image(rgb_img, grayscale_cam, use_rgb=True)

            outputs = model(input_tensor)
            _, pred = torch.max(outputs, 1)

            plt.subplot(num_images//2, 4, images_so_far*2+1)
            plt.imshow(rgb_img)
            plt.title(f'True: {classes[labels[i]]}')
            plt.axis('off')

            plt.subplot(num_images//2, 4, images_so_far*2+2)
            plt.imshow(visualization)
            plt.title(f'Pred: {classes[pred[0]]}')
            plt.axis('off')

            images_so_far += 1

visualize_grad_cam(model, test_loader, classes)
plt.tight_layout()
plt.show()
```

```python
# Extract and print overall accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f"Overall Accuracy: {accuracy:.4f}")
```
